```python
def get_data_ready(num_features = 9):
#     start_time = time.time()
#     num_features, C = find_optimal_values(max_features=15, max_neighbours=31, num_splits=10, symbol_name = 'MSFT')
#     end_time = time.time()
#     print('Time taken:', end_time - start_time)
    
    X_train, X_test, Y_train, Y_test = hm.prepare_data(num_features)
    X_train, X_test, Y_train, Y_test = X_train.values, X_test.values, Y_train.values, Y_test.values
    return X_train, X_test, Y_train, Y_test  
```

### 1. SkLearn RF Classifier

```python
def sklearn_RF_forecast(X_train, X_test, Y_train, Y_test):
    for n_estimators in list(range(100, 1100, 100)):
        clf = RandomForestClassifier(n_estimators=n_estimators, n_jobs=-1)
        clf.fit(X_train, Y_train)
        
        print('In-Built SVM (Accuracy) score -- n_estimators =', n_estimators, '--', clf.score(X_test, Y_test), '\n')
```

### 2. Predicting using Implementation

### Running RF

```python
def forecast(X_train, X_test, Y_train, Y_test):
    sklearn_RF_forecast(X_train, X_test, Y_train, Y_test)
    
# X_train, X_test, Y_train, Y_test = hm.prepare_data(2)
# X_train, X_test, Y_train, Y_test = X_train.values, X_test.values, Y_train.values, Y_test.values
# forecast(X_train, X_test, Y_train, Y_test)
```

```python
def run_RF():
    for num_features in range(1, 16):
        print('Features -------------------------------->', num_features)
        X_train, X_test, Y_train, Y_test = get_data_ready(num_features = num_features)
        forecast(X_train, X_test, Y_train, Y_test)
```
